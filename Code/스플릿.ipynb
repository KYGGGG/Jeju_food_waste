```python
import pandas as pd
from statsmodels.formula.api import ols

f1 = 'em_t ~ base_month + area + em_area_cd_cnt + emd_nm + resid_reg_pop + foreign_pop + d_visit_pop_cnt + l_visit_pop_cnt + s_visit_pop_cnt + pre_mm'
multi_model = ols(formula = f1, data = df).fit()
multi_model.summary()
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

fitted = multi_model.predict(df)
residual = df['em_t'] - fitted

sns.regplot(fitted, residual, lowess=True, line_kws={'color': 'red'})
plt.plot([fitted.min(), fitted.max()], [0, 0], '--', color='grey')
```

```python
import numpy as np
sns.regplot(fitted, np.sqrt(np.abs(multi_model)), lowess=True, line_kws={'color': 'red'})
```
